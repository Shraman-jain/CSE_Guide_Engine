```python
import requests
from bs4 import BeautifulSoup

def scrapping(url,name):
  # Send a GET request to the website
  #url = "https://www.geeksforgeeks.org/python-lists/"
  response = requests.get(url)

  # Parse the HTML content using Beautiful Soup
  soup = BeautifulSoup(response.content, 'html.parser')

  # Find the first <h1> tag in the HTML content
  paragraphs = soup.find_all('p')
  headings = soup.find_all('h1')
  #ul_lists = soup.find_all('ul')
  #li_items = soup.find_all('li')

  text1 = ""
  for tag in headings + paragraphs:
      tag_text = tag.text.strip()
      text1 = text1+"\n"+tag_text+"\n"


  with open(name, 'w') as file:
      file.write(text1)
      file.close()
```

```python
def removingdigits(example):
  import re
  # Read the text file into a string variable
  with open(example, 'r') as file:
      paragraph = file.read()

  # Remove numbers from paragraph
  clean_paragraph = ''.join(filter(lambda x: not x.isdigit(), paragraph))
  # Write the cleaned paragraph back to the text file
  with open(example, 'w') as file:
      file.write(clean_paragraph)
```

```python
#Doing lower case 

def lowercase(example):
  file1 = open(example,"r+") 
  contents = file1.read()
  contents = contents.lower()
  file1.close()
  file1 = open(example, "w")
  file1.seek(0)
  file1.write(contents)
  file1.truncate()
  file1.close()
```

```python
#Stopwords

def stopwords(example):
  import nltk
  nltk.download('punkt')
  import string
  from bs4 import BeautifulSoup
  import os
  nltk.download('stopwords')
  from nltk.corpus import stopwords
  stopword1=stopwords.words('english')

  j=0 
  file_content = open(example,"r+")
  contents=file_content.read()
  tokens=nltk.word_tokenize(contents)
  contents = [w for w in tokens if not w in stopword1]
  str_wt_s=""
  for t in contents:
    str_wt_s+= t+ " "
  file_content.close()
  file1 = open(example, "w")
  file1.seek(0)
  file1.write(str_wt_s)
  file1.truncate()
  file1.close()
```

```python
def punctuations(example):
#Punctuations
  punc = '''!()-[]’'“|”–{}./;:'"\,<>./?@#$%^&*_~`'''

  file1=open(example,'r')
  line = file1.read()
  words = line.split(" ")
  file1.close()
  temp = []
  for r in words:
    if r not in punc:
      temp.append(r)

  file = open(example,'w')
  for item in temp:
    file.write(item+" ")
  file.close()
```

```python
def blankspace(example):
#For removing blank space
  j=0    
  file1 = open(example,"r+") 
  without_blank = ""
  for j in file1:
    if not j.isspace():
        without_blank+=j

  file1.close()
  file1 = open(example, "w")
  file1.seek(0)
  file1.write(without_blank)
  file1.truncate()
  file1.close()
```

```python
#preprocess
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~`'''
stop_words = set(stopwords.words('english'))
def preprocess(contents):
    #lowercase
    contents_lower = contents.lower()
    #tokenize
    contents_token = nltk.word_tokenize(contents_lower)
    temp = []
    #remove stopwords
    for r in contents_token:
        if r not in stop_words:
            temp.append(r)
    contents_token = ""
    for i in temp:
        contents_token = contents_token+" "+i
    # remove punctuation
    no_punc = ""
    for char in contents_token:
        if char not in punc:
            no_punc = no_punc + char
    # removing spaces
    res = re.sub(' +', ' ', no_punc)
    res = res.strip()
    return str(res)
```
